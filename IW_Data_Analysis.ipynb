```
# load county populations
popDF = pd.read_csv("/content/drive/MyDrive/Junior Year/IW 07/2016PopEstimates.csv")
GDF14 = GDF14.assign(POPINC=pd.Series(np.zeros(len(GDF14))).values)
GDF14 = GDF14.assign(POP=pd.Series(np.zeros(len(GDF14))).values)

popDF.set_index('GEOID', inplace=True)
popDF.head()
```

```
# calculate incidents scaled by population
for i in GDF14.index:
  try:
    GDF14.loc[i, 'POPINC'] = GDF14.loc[i, 'numIncidents'] / popDF.loc[int(i), 'POPESTIMATE2016']
    GDF14.loc[i, 'POP'] = popDF.loc[int(i), 'POPESTIMATE2016']
  except KeyError:
    pass
  
GDF14.nlargest(5, 'POPINC')
```
